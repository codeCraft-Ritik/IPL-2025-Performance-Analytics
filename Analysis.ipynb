```python
import pandas as pd

batters = pd.read_csv("IPL2025Batters.csv")
bowlers = pd.read_csv("IPL2025Bowlers.csv")
```

```python
print(batters.head())
```

```python
print(bowlers.head())
```

## Top Run Scorers in IPL 2025

```python
top_runs = batters.sort_values(by="Runs", ascending=False).head(10)
print("\n" + "="*80)
print("TOP 10 RUN SCORERS IN IPL 2025")
print("="*80 + "\n")
print(top_runs[["Player Name", "Team", "Runs", "Matches", "AVG", "SR"]].to_string(index=False))
print("\n" + "="*80 + "\n")
```

## Best Strike Rate

```python
best_sr = batters.sort_values(by="SR", ascending=False).head(10)
print("\n" + "="*80)
print("TOP 10 BATTERS BY STRIKE RATE")
print("="*80 + "\n")
print(best_sr[["Player Name", "Team", "Runs", "SR", "Matches"]].to_string(index=False))
print("\n" + "="*80 + "\n")
```

## Chart 

```python
import matplotlib.pyplot as plt

# Visualization for Top Run Scorers
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Top Runs Bar Chart
top_runs_chart = batters.sort_values(by="Runs", ascending=False).head(10)
axes[0].barh(top_runs_chart["Player Name"], top_runs_chart["Runs"], color='steelblue')
axes[0].set_xlabel("Runs Scored", fontsize=12, fontweight='bold')
axes[0].set_title("Top 10 Run Scorers in IPL 2025", fontsize=14, fontweight='bold')
axes[0].invert_yaxis()

for i, v in enumerate(top_runs_chart["Runs"]):
    axes[0].text(v + 5, i, str(int(v)), va='center', fontweight='bold')

# Best Strike Rate Bar Chart
best_sr_chart = batters.sort_values(by="SR", ascending=False).head(10)
axes[1].barh(best_sr_chart["Player Name"], best_sr_chart["SR"], color='coral')
axes[1].set_xlabel("Strike Rate (%)", fontsize=12, fontweight='bold')
axes[1].set_title("Top 10 Batters by Strike Rate", fontsize=14, fontweight='bold')
axes[1].invert_yaxis()

for i, v in enumerate(best_sr_chart["SR"]):
    axes[1].text(float(v) + 2, i, f"{v}", va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

## Bowler Analysis

Top Wicket Takers

```python
print("\n" + "="*100)
print("ALL BOWLERS IN IPL 2025 - COMPLETE LIST")
print("="*100 + "\n")
print(bowlers[["Player Name", "Team", "WKT", "MAT", "AVG", "ECO", "SR"]].to_string(index=False))
print("\n" + "="*100 + "\n")
```

```python
top_wickets = bowlers.sort_values(by="WKT", ascending=False).head(10)
print("\n" + "="*100)
print("TOP 10 WICKET TAKERS IN IPL 2025")
print("="*100 + "\n")
print(top_wickets[["Player Name", "Team", "WKT", "MAT", "OVR", "AVG", "ECO", "SR"]].to_string(index=False))
print("\n" + "="*100 + "\n")
```

```python
# Visualization for Top Wicket Takers
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Top Wickets Bar Chart
top_wickets_chart = bowlers.sort_values(by="WKT", ascending=False).head(10)
axes[0].barh(top_wickets_chart["Player Name"], top_wickets_chart["WKT"], color='darkgreen')
axes[0].set_xlabel("Wickets Taken", fontsize=12, fontweight='bold')
axes[0].set_title("Top 10 Wicket Takers in IPL 2025", fontsize=14, fontweight='bold')
axes[0].invert_yaxis()

for i, v in enumerate(top_wickets_chart["WKT"]):
    axes[0].text(v + 0.5, i, str(int(v)), va='center', fontweight='bold')

# Best Economy Rate Bar Chart (Lower is better)
best_eco = bowlers.sort_values(by="ECO", ascending=True).head(10)
axes[1].barh(best_eco["Player Name"], best_eco["ECO"], color='purple')
axes[1].set_xlabel("Economy Rate (Runs per Over)", fontsize=12, fontweight='bold')
axes[1].set_title("Top 10 Bowlers by Best Economy Rate", fontsize=14, fontweight='bold')
axes[1].invert_yaxis()

for i, v in enumerate(best_eco["ECO"]):
    axes[1].text(float(v) + 0.1, i, f"{v}", va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

## Visualization

```python
import seaborn as sns

# Create a professional seaborn barplot
fig, ax = plt.subplots(figsize=(12, 8))
top_runs_data = batters.sort_values(by="Runs", ascending=True).tail(10)
sns.barplot(data=top_runs_data, x="Runs", y="Player Name", palette="Blues_r", ax=ax)
ax.set_title("Top 10 Run Scorers in IPL 2025", fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel("Total Runs Scored", fontsize=12, fontweight='bold')
ax.set_ylabel("Player Name", fontsize=12, fontweight='bold')

for i, v in enumerate(top_runs_data["Runs"]):
    ax.text(v + 10, i, str(int(v)), va='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

## Find Best All-Rounders

```python
all_rounders = pd.merge(batters[["Player Name", "Team", "Runs", "Matches", "AVG", "SR"]], 
                         bowlers[["Player Name", "Team", "WKT", "MAT", "AVG", "ECO"]], 
                         on="Player Name", 
                         suffixes=("_bat", "_bowl"),
                         how="inner")

all_rounders["All_Rounder_Score"] = (
    all_rounders["Runs"] * 0.5 +
    all_rounders["WKT"] * 25 -
    all_rounders["ECO"] * 2
)

best_allrounders = all_rounders.sort_values(by="All_Rounder_Score", ascending=False).head(10)

print("\n" + "="*120)
print("TOP ALL-ROUNDERS IN IPL 2025 (Based on Batting + Bowling Performance)")
print("="*120 + "\n")
print(best_allrounders[["Player Name", "Team_bat", "Runs", "WKT", "AVG_bat", "ECO", "All_Rounder_Score"]].to_string(index=False))
print("\n" + "="*120 + "\n")
```

```python
# Visualization for All-Rounders
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# All-Rounder Score Bar Chart
axes[0].barh(best_allrounders["Player Name"], best_allrounders["All_Rounder_Score"], color='gold')
axes[0].set_xlabel("All-Rounder Score", fontsize=12, fontweight='bold')
axes[0].set_title("Top 10 All-Rounders in IPL 2025", fontsize=14, fontweight='bold')
axes[0].invert_yaxis()

for i, v in enumerate(best_allrounders["All_Rounder_Score"]):
    axes[0].text(v + 1, i, f"{v:.1f}", va='center', fontweight='bold')

# Scatter plot: Runs vs Wickets
scatter = axes[1].scatter(best_allrounders["Runs"], best_allrounders["WKT"], 
                        s=best_allrounders["All_Rounder_Score"]*5, alpha=0.6, c=best_allrounders["All_Rounder_Score"], 
                        cmap='YlOrRd', edgecolors='black', linewidth=1.5)
axes[1].set_xlabel("Runs Scored", fontsize=12, fontweight='bold')
axes[1].set_ylabel("Wickets Taken", fontsize=12, fontweight='bold')
axes[1].set_title("All-Rounders: Batting vs Bowling Performance", fontsize=14, fontweight='bold')

for idx, row in best_allrounders.iterrows():
    axes[1].text(row["Runs"], row["WKT"], row["Player Name"], fontsize=9, ha='center', va='center', fontweight='bold')

plt.colorbar(scatter, ax=axes[1], label='All-Rounder Score')
plt.tight_layout()
plt.show()
```
